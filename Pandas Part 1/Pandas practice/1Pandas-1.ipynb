### üìå 1. Data Creation & Importing
pd.DataFrame(data, columns=[...]) ‚Üí Creates a DataFrame from a dictionary, list, or NumPy array.

pd.Series(data, index=[...]) ‚Üí Creates a Pandas Series (1D labeled array).

pd.read_csv('file.csv') ‚Üí Reads a CSV file into a DataFrame.

pd.read_excel('file.xlsx', sheet_name='Sheet1') ‚Üí Reads an Excel file.

pd.read_json('file.json') ‚Üí Reads a JSON file.

pd.read_sql(query, connection) ‚Üí Reads data from a SQL database.

### üìå 2. Basic DataFrame Operations
df.head(n=5) ‚Üí Returns the first n rows.

df.tail(n=5) ‚Üí Returns the last n rows.

df.info() ‚Üí Displays information about the DataFrame.

df.describe() ‚Üí Provides summary statistics.

df.shape ‚Üí Returns the dimensions of the DataFrame.

df.columns ‚Üí Lists all column names.

df.dtypes ‚Üí Displays data types of columns.

### üìå 3. Selection & Filtering
df['column_name'] ‚Üí Selects a single column.

df[['column1', 'column2']] ‚Üí Selects multiple columns.

df.iloc[2, 3] ‚Üí Selects rows/columns by position (index-based).

df.loc[2, 'col'] ‚Üí Selects rows/columns by label.

df[df['Salary'] > 50000] ‚Üí Filters rows based on conditions.

df.query("Salary > 50000") ‚Üí Queries data using string conditions.

### üìå 4. Data Cleaning & Handling Missing Values
df.dropna() ‚Üí Drops rows with missing values.

df.fillna(value) ‚Üí Replaces missing values with a specified value.

df.isna() ‚Üí Checks for missing values (returns True/False).

df.duplicated() ‚Üí Checks for duplicate rows.

df.drop_duplicates() ‚Üí Removes duplicate rows.

### üìå 5. Data Transformation
df['col'].apply(func) ‚Üí Applies a function to a column.

df['col'].map(lambda x: x*2) ‚Üí Maps a function element-wise.

df['col'].replace(old, new) ‚Üí Replaces specific values.

df.rename(columns={'old': 'new'}) ‚Üí Renames columns.

df['col'].astype('int') ‚Üí Converts data types.

### üìå 6. Sorting & Ordering
df.sort_values(by='col', ascending=False) ‚Üí Sorts by column values.

df.sort_index(ascending=False) ‚Üí Sorts by index.

### üìå 7. Aggregation & Grouping
df.groupby('col').mean() ‚Üí Groups data by column and aggregates.

df.pivot(index='col1', columns='col2', values='col3') ‚Üí Reshapes data without aggregation.

df.pivot_table(values='col', index='col1', aggfunc='sum') ‚Üí Aggregates and reshapes data.

df['col'].value_counts() ‚Üí Counts unique values.

### üìå 8. Merging & Joining
pd.concat([df1, df2], axis=0) ‚Üí Combines DataFrames vertically or horizontally.

df1.merge(df2, on='col', how='inner') ‚Üí Merges DataFrames on a key column.

df1.join(df2, on='key') ‚Üí Joins DataFrames based on the index.

### üìå 9. Window Functions
df['col'].rolling(3).mean() ‚Üí Computes moving averages.

df['col'].expanding().sum() ‚Üí Expanding window calculation.

df['col'].cumsum() ‚Üí Cumulative sum.

df['col'].cumprod() ‚Üí Cumulative product.

### üìå 10. Exporting Data
df.to_csv('file.csv', index=False) ‚Üí Saves DataFrame to a CSV file.

df.to_excel('file.xlsx', sheet_name='Sheet1') ‚Üí Saves DataFrame to an Excel file.

df.to_json('file.json') ‚Üí Saves DataFrame as a JSON file.

df.to_sql('table_name', connection, if_exists='replace') ‚Üí Saves DataFrame to a SQL table.
