```python
import re
corpus=re.sub("[^A-Za-z0-9 ]","",corpus)
print(corpus)
```

```python
token=word_tokenize(corpus)
lower=[]
for i in token:
    lower.append(i.lower())
print(lower)
```

```python
stop=[i for i in lower if i not in stopwords.words("english")]
print(stop)
```

```python
lemma=WordNetLemmatizer()
words=[lemma.lemmatize(i) for i in stop]
```

```python
print(words[:50])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def my_tokenizer(text):
    return text.lower().split()
```

```python
count=CountVectorizer(
    decode_error='strict',
    lowercase=True,
    tokenizer=my_tokenizer, token_pattern=None,
    stop_words='english',
    ngram_range=(1, 1),
    max_df=1,
    min_df=0.0,
    max_features=100,
    binary=False)
val=count.fit_transform(corpus.split(" "))
val.toarray()
```

```python
vector=CountVectorizer(max_features=100,)
vectors=vector.fit_transform(words)
arr=vectors.toarray()
arr
```

```python
feature_names=vector.get_feature_names_out()
feature_names
```
