```python
# recursion
def fact(n):
    if n==1:
        return 1
    elif n>1:
        return (n)*fact(n-1)
fact(5)
```

```python
def fibb(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    elif n==2:
        return 1
    else:
        return fibb(n-1)+fibb(n-2)
for i in range(10):
    print(fibb(i),end=" ")
```

```python
def process_queries(n, arr, queries):
    results = []
    for query in queries:
        query_type = query[0]
        if query_type == 1:
            l, x = query[1], query[2]
            arr[l-1] += x
        elif query_type == 2:
            l, r, x = query[1], query[2], query[3]
            count = 0
            for i in range(l-1, r):
                if arr[i] > x:
                    count += 1
            results.append(count)
    return results

n = int(input())
arr = list(map(int, input().split()))
q = int(input())
queries = []
for _ in range(q):
    queries.append(list(map(int, input().split())))
output = process_queries(n, arr, queries)
for result in output:
    print(result)
```

```python
def solve(N, A, Q, queries):
    results = []
    for query in queries:
        if query[0] == 1:
            L, X = query[1], query[2]
            A[L - 1] = X
        elif query[0] == 2:
            L, R, X = query[1], query[2], query[3]
            found = False
            for i in range(L, R):
                if A[i - 1] < X:
                    results.append(i)
                    found = True
                    break
            if not found:
                results.append(-1)
    return results

N = int(input())
A = list(map(int, input().split()))
Q = int(input())
queries = []
for _ in range(Q):
    queries.append(list(map(int, input().split())))
output = solve(N, A, Q, queries)
for result in output:
    print(result)
```
