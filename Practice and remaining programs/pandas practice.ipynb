```python
dataframe={"Emp id":np.random.randint(1,100,50),
           "Age":np.random.randint(1,100,50),
           "Salary":np.random.randint(50000,70000,50)}
df=pd.DataFrame(dataframe)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# fig,axes=plt.subplots(2,2,dpi=100)
sns.barplot(x=df["Age"],y=df["Salary"],hue=df["Age"])
plt.tight_layout()
plt.xticks(rotation=90)
plt.show()
```
