Here is a list of **Matplotlib built-in functions** in order, starting from **figure creation** to **plot customization**, **annotations**, and **saving the plot**.  

---

### **1. Figure and Axes Management**  
These functions initialize and manage figures and subplots.  
- `plt.figure(figsize=(width, height), dpi=dpi_value)`: Creates a new figure with a specified size and resolution.  
- `fig, ax = plt.subplots(nrows, ncols, figsize=(width, height))`: Creates multiple subplots in a figure.  
- `plt.subplot(nrows, ncols, index)`: Creates a subplot in a grid layout.  
- `plt.tight_layout()`: Adjusts subplot spacing to prevent overlap.  

---

### **2. Plotting Functions**  
These functions generate different types of plots.  
- `plt.plot(x, y, linestyle='', marker='', color='', label='')`: Creates a line plot.  
- `plt.scatter(x, y, s=marker_size, c=color, label='', alpha=opacity)`: Creates a scatter plot.  
- `plt.bar(x, height, width=bar_width, color='', label='', align='')`: Creates a bar chart.  
- `plt.barh(y, width, height=bar_height, color='', label='')`: Creates a horizontal bar chart.  
- `plt.hist(data, bins=number_of_bins, color='', edgecolor='', alpha=opacity)`: Plots a histogram.  
- `plt.boxplot(data, notch=False, vert=True, patch_artist=True)`: Creates a box plot.  
- `plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, explode=explode_values, shadow=True)`: Draws a pie chart.  

---

### **3. Axis and Grid Customization**  
These functions modify the axes and add grid lines.  
- `plt.xlabel('X-axis label', fontsize=size, color=color)`: Sets the x-axis label.  
- `plt.ylabel('Y-axis label', fontsize=size, color=color)`: Sets the y-axis label.  
- `plt.title('Graph Title', fontsize=size, color=color, loc='left/right/center')`: Adds a title to the plot.  
- `plt.legend(loc='upper/lower left/right', fontsize=size, title='Legend Title')`: Adds a legend.  
- `plt.grid(True, linestyle='--', linewidth=width, alpha=opacity)`: Adds grid lines.  
- `plt.xticks(ticks, labels, rotation=angle, fontsize=size)`: Customizes x-axis tick labels.  
- `plt.yticks(ticks, labels, rotation=angle, fontsize=size)`: Customizes y-axis tick labels.  

---

### **4. Axis Limits and Scaling**  
These functions adjust axis ranges and scales.  
- `plt.xlim(left=min_value, right=max_value)`: Sets x-axis limits.  
- `plt.ylim(bottom=min_value, top=max_value)`: Sets y-axis limits.  
- `plt.axis([xmin, xmax, ymin, ymax])`: Sets limits for both x and y axes.  
- `plt.xscale('linear/log/symlog/logit')`: Changes x-axis scale.  
- `plt.yscale('linear/log/symlog/logit')`: Changes y-axis scale.  

---

### **5. Annotations and Text**  
These functions help in labeling data points.  
- `plt.text(x, y, 'Text', fontsize=size, color=color, ha='center/left/right', va='center/top/bottom')`: Adds text at a specific position.  
- `plt.annotate('Annotation', xy=(x, y), xytext=(x_offset, y_offset), arrowprops={'arrowstyle': '->'})`: Adds an annotation with an arrow.  

---

### **6. Saving and Displaying the Plot**  
These functions control how the plot is saved or displayed.  
- `plt.savefig('filename.png', dpi=resolution, format='png/pdf/svg', bbox_inches='tight')`: Saves the figure.  
- `plt.show()`: Displays the plot.  

---
