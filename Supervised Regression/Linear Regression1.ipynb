```python
import numpy as np
import pandas as pd
import scipy as sci
import matplotlib.pyplot as plt
import seaborn as sns
# from sklearn.datasets import load_iris
```

```python
df=pd.read_csv("https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv")
```

```python
df.info()
```

```python
# sns.(x=df["Salary"],y=df["Experience Years"])
sns.pairplot(data=df)
plt.show()
```

```python
x=df[["Experience Years"]]
y=df["Salary"]
x.head(10)
```

```python
y.head(10)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
linear=LinearRegression()
linear.fit(x_train,y_train)
```

```python
linear.intercept_
```

```python
linear.coef_
```

```python
y_pred=linear.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
rmse=np.sqrt(mse)
rmse
```

```python
r2_score(y_test,y_pred)
```

```python
fig=plt.figure(dpi=200)
plt.scatter(x_test,y_test,label="Actual",color="violet")
plt.plot(x_test,y_pred,label="Predict",linewidth=2,color="Green")
plt.xlabel("Years of experience")
plt.ylabel("Salary of the employees")
plt.title("The linear regression of the predicted salaries")
plt.show()
```
